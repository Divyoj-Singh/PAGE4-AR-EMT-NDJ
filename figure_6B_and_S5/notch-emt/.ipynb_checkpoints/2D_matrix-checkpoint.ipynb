```python
import auxilary_functions as aux
import numpy as np

%load_ext autoreload
%autoreload 2
%matplotlib inline

p   = aux.parameters()
eqs = aux.equations()

# Figure S3
p['n']  = 50     # number of cells = n x n
p['dt'] = 0.1    # time step of the Euler method

key = 'W'                  # key parameter for hexagonal snapshot plot
clim = [0,40000]           # Color range for the plot
tr  = {'W': [5000, 15000]}

p['t']  = p['dt']
pts_i = aux.euler_traj(eqs, p, pts=None, vlim=aux.vlim())
aux.plot_hex(pts_i[key], clim=clim, clabel=key)
aux.plot_hex(pts_i[key], clim=clim, tr=[5000,15000], c=[34000, 20000, 10000], cbar=False)

p['t']  = 120   # time in hours 
```

```python
# Figures 4A and 4C
p['gJ'] = 20
v = 'gD'
r_v = np.arange(10, 201, 10)
aux.plot_fractionStates(eqs, p, v, r_v, key, tr, pts_i=pts_i, vlim=aux.vlim(), show_snapshot=False)

p['gD'] = 70    
pts_D070 = aux.euler_traj(eqs, p, pts=pts_i, vlim=aux.vlim())
aux.plot_hex(pts_D070[key], clim=clim, tr=[5000,15000], c=[34000, 22000, 10000], cbar=False)

# Figures 4B and 4D
p['gD'] = 20
v = 'gJ'
r_v = np.arange(10, 201, 10)
aux.plot_fractionStates(eqs, p, v, r_v, key, tr, pts_i=pts_i, vlim=aux.vlim(), show_snapshot=False)

#p['gJ'] = 70    
# Different value for snapshot illustration
p['gJ'] = 40    
pts_J070 = aux.euler_traj(eqs, p, pts=pts_i, vlim=aux.vlim())
aux.plot_hex(pts_J070[key], clim=clim, tr=[5000,15000], c=[34000, 22000, 10000], cbar=False)
```
